Sales and Profit Analysis 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv('StoreUS-2015.csv')

data['Order Date'] = pd.to_datetime(data['Order Date'], errors='coerce')
data['Ship Date'] = pd.to_datetime(data['Ship Date'], errors='coerce')

# Calculate total sales and profit by region
sales_profit_by_region = data.groupby('Region').agg({
    'Sales': 'sum',
    'Profit': 'sum'
}).reset_index()

# Calculate total sales and profit by state
sales_profit_by_state = data.groupby('State or Province').agg({
    'Sales': 'sum',
    'Profit': 'sum'
}).reset_index()

# Calculate total sales and profit by product category
sales_profit_by_category = data.groupby('Product Category').agg({
    'Sales': 'sum',
    'Profit': 'sum'
}).reset_index()

# Step 5: Visualize Sales and Profit Trends Over Time
sales_profit_over_time = data.groupby('Order Date').agg({
    'Sales': 'sum',
    'Profit': 'sum'
}).reset_index()

#Plotting Sales and Profit by Region
plt.figure(figsize=(10, 5))
plt.bar(sales_profit_by_region['Region'], sales_profit_by_region['Sales'], color='lightblue', label='Sales')
plt.bar(sales_profit_by_region['Region'], sales_profit_by_region['Profit'], color='orange', label='Profit')
plt.title('Total Sales and Profit by Region')
plt.xlabel('Region')
plt.ylabel('Amount in $')
plt.legend()
plt.show()

# Ploting Sales and Profit by State
plt.figure(figsize=(10, 5))
states = sales_profit_by_state['State or Province']
plt.bar(states, sales_profit_by_state['Sales'], color='green', label='Sales')
plt.bar(states, sales_profit_by_state['Profit'], color='red', label='Profit')
plt.title('Total Sales and Profit by State')
plt.xlabel('State')
plt.ylabel('Amount in $')
plt.xticks(rotation=90)
plt.legend()
plt.show()

# Plot Sales and Profit by Product Category
plt.figure(figsize=(10, 5))
categories = sales_profit_by_category['Product Category']
plt.bar(categories, sales_profit_by_category['Sales'], color='purple', label='Sales')
plt.bar(categories, sales_profit_by_category['Profit'], color='yellow', label='Profit')
plt.title('Total Sales and Profit by Product Category')
plt.xlabel('Product Category')
plt.ylabel('Amount in $')
plt.legend()
plt.show()

# Ploting Sales and Profit Trends Over Time
plt.figure(figsize=(10, 5))
plt.plot(sales_profit_over_time['Order Date'], sales_profit_over_time['Sales'], color='blue', label='Sales')
plt.plot(sales_profit_over_time['Order Date'], sales_profit_over_time['Profit'], color='orange', label='Profit')
plt.title('Sales and Profit Trends Over Time')
plt.xlabel('Order Date')
plt.ylabel('Amount in $')
plt.legend()
plt.grid(True)
plt.show()
```
